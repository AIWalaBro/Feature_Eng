```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

# missing value checking

```python
df.isnull().sum()
```

# lets fill the misiing values

```python
# simple imputer creating object

si_imp_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

```

```python
# simple imputer in age will fill misiing values with mean

x_train_age = si_imp_age.fit_transform(x_train[['Age']])
x_train_embarked = si_embarked.fit_transform(x_train[['Embarked']])
```

```python
# simple imputer in Embarked will be fill with most frequent

x_test_age = si_imp_age.transform(x_test[['Age']])
x_test_embarked = si_embarked.transform(x_test[['Embarked']])
```

```python
# lest check out are the misiing values fill with null or not
x_train_embarked
```

```python
x_test_age
```

# One hot encoding


```python
# tum alag object create krke embarked aur sex ke upar one hot encode q perform kr rhe ho?
# ans:- kyoki esme missing values hai embarked vale column me 
# but aab puchoge ki last avle esme hi toh embarked vale column me perform kiya hai na?
# ans:- original vale me koe changes nhi hai voh object crete kiya hai humne uska 
# sparse by default true rhta hai

# sparse by defalult true rhta hai and if true rha toh sparse metrics will be created 
# and handle_unknown = 'ignore' mean age future me values create hota hai toh use voh 0 creeate krega

ohe_sex = OneHotEncoder(sparse = False,handle_unknown = 'ignore')
ohe_embarked = OneHotEncoder(sparse = False,handle_unknown = 'ignore')

# x_train me hum fit transform krte hai
x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(x_train_embarked)

# x_test me sirf transform krte hai
x_test_sex = ohe_sex.transform(x_test[['Sex']])
x_test_embarked = ohe_embarked.transform(x_test_embarked)





```

```python
x_train_sex
# hum 'drop = first' q nhi liye kyoki hum d.t algorithm used krne wale hai so multicolinearity 
# koe problem nhi uspe effect nhi krta linear li solve nhi kr rhe hai hum.
```

```python
x_train_embarked
```

```python
# we dropping the column of sex,age,embarked why?
# ans:- because we alredy transform the columns into missing into the simple imputer and embarked converted 
#      into the one hot encode taht why

x_train_remain = x_train.drop(columns=['Sex','Age','Embarked'])
x_test_remain  = x_test.drop(columns=['Sex','Age','Embarked'])
```

```python
# concatenate all the array that we get

x_train_transformed = np.concatenate((x_train_remain,x_train_embarked,x_train_age,x_train_sex), axis= 1)
x_test_transformed  = np.concatenate((x_test_remain,x_test_embarked,x_test_age,x_test_sex), axis= 1)
```

# lets create the decision tree model


```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed,y_train)
```

# lets predict the test model

```python
y_pred = clf.predict(x_test_transformed)
```

```python
y_pred
```

# lets see the accuracy 

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
